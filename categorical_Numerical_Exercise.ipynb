# using One Hot Encoder

```python
df2 = pd.read_csv("/content/carprices.csv")
```

```python
df2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df2["Car Model"] = le.fit_transform(df2["Car Model"])
```

```python
df2
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([("Car Model" , OneHotEncoder() , [0])] , remainder = "passthrough")
```

```python
X = df2[["Car Model"]].values
```

```python
t = ct.fit_transform(X)
```

```python
t
```

```python
tdf = pd.DataFrame(t)
```

```python
final = pd.concat([df2,tdf] , axis = "columns")
```

```python
final
```

```python
final.drop(["Car Model" , 2] , axis = "columns")
```

```python
yfinal = final["Sell Price($)"]
```

```python
yfinal
```

```python
xfinal = final.drop(["Sell Price($)" , "Car Model"] , axis = "columns")
```

```python
xfinal
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model2 = LinearRegression()
```

```python
xfinal.columns = xfinal.columns.astype(str)
```

```python
xfinal
```

```python
xfinal.drop("2" , axis = "columns", inplace = True)
```

```python
model2.fit(xfinal , yfinal)
```

```python
model2.score(xfinal,yfinal)
```

```python
model2.predict([[45000,4,0,0]])
```

```python
model2.predict([[86000,7,0,1]])
```
