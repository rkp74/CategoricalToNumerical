```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hp = pd.read_csv("/content/homeprices.csv")
```

# Use Get Dummies


```python
hp.head(5)
```

```python
dummies = pd.get_dummies(hp.town)
```

```python
merged = pd.concat([hp,dummies] , axis = 'columns')
```

```python
merged
```

```python
final = merged.drop(['town' , 'west windsor'] , axis = "columns")
```

```python
final
```

```python
X = final.drop('price'  , axis = "columns")
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,final.price)
```

```python
model.predict([[3600,1,0]]) # monroe township
```

```python
model.predict([[2600,0,1]]) # robinsville
```

```python
model.predict([[3400,0,0]]) # west windsor
```

```python
model.score(X,final.price)
```

# using One Hot Encoding

```python
df = hp
```

```python
df.head
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.town = le.fit_transform(df.town)
```

```python
df
```

```python
X = df[['town' , 'area']].values
X
```

```python
y = df.price
y
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([("town" , OneHotEncoder() , [0])] , remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
X = X[: , 1:]
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[0,0,3600]]) # monroe township
```

```python
model.predict([[0,1,3400]]) # west windsor
```

```python
model.predict([[1,0,2600]]) # robinsville
```
